# Query

- .query() är mycket mer lättläst än att göra en *mask*

```python
salaries = pd.read_csv("../data/salaries.csv")
salaries.head()
```

```python
salaries[salaries["job_title"] == "ML Engineer"]
```

```python
sns.relplot(x="salary_in_usd", y="company_location", data=salaries[salaries["job_title"] == "ML Engineer"])
sns.relplot(x="salary_in_usd", y="company_location", data=salaries.query("job_title == 'ML Engineer'"))
```

```python
salary = salaries[salaries["salary_in_usd"] > 400000]
salary
```

```python
salaries.query("`salary_in_usd` and `salary` > 400000")
```

```python
sal_copy = salaries.copy()

sal_copy.loc[sal_copy["salary_in_usd"] > 400000, ["salary"]]
```

## Sortera DataFrames

- Använd sort_values()!
- Parameters:
    - **by** kolumner, lista av kolumner
    - **ascending** Valbart, Default=True

```python
salaries.sort_values(by="salary_in_usd", ascending=False)
```

```python
salaries.sort_values(by="salary", ascending=False)
```

```python
salaries.sort_values(by=["salary", "salary_in_usd"], ascending=[True, False])
```
