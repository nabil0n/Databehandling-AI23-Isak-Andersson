```python
import seaborn as sns

tips = sns.load_dataset('tips')
tips
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue="sex", style="smoker", size="size", col="day", height=2.5, col_wrap=2, palette="Set2", kind="scatter", s=100)
```

```python
sns.displot(x='total_bill', y='tip', data=tips)
```
